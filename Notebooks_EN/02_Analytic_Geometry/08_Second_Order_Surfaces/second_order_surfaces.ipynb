## 17. Equations of second-order surfaces

* Write the equation of a sphere with center at point $P=(1,2,3)$ and radius $r=3$.
* Do the spheres with equations $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$ have any common points?
* What curve in space is formed by the intersection of the sphere $x^2 + y^2 + z^2 = 1$ with the sphere $(x-1)^2 + y^2 + z^2 = 1$? Find the equation of this curve.
* Write the equation of the tangent plane to the paraboloid $z=(x-1)^2+y^2+1$ at point $P(1,0,1)$.

### 1. Write the equation of a sphere with center at point \( P = (1, 2, 3) \) and radius \( r = 3 \).

### Solution:

The general equation of a sphere with center at \( (x_0, y_0, z_0) \) and radius \( r \) is:

$$
(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2
$$

Substituting the given values \( P = (1, 2, 3) \) and \( r = 3 \), we get:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 3^2
$$

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

Thus, the equation of the sphere is:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the sphere parameters
center = np.array([1, 2, 3])  # Center at (1, 2, 3)
radius = 3  # Radius = 3

# Create a meshgrid of spherical coordinates (theta and phi)
theta = np.linspace(0, np.pi, 100)  # Polar angle
phi = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle

# Create a grid of points in spherical coordinates
theta, phi = np.meshgrid(theta, phi)

# Parametrize the sphere in Cartesian coordinates
x = center[0] + radius * np.sin(theta) * np.cos(phi)
y = center[1] + radius * np.sin(theta) * np.sin(phi)
z = center[2] + radius * np.cos(theta)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the sphere
ax.plot_surface(x, y, z, color='cyan', alpha=0.6)

# Plot the center of the sphere
ax.scatter(center[0], center[1], center[2], color='red', s=100, label="Center P(1, 2, 3)")

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Sphere with Center (1, 2, 3) and Radius 3')

# Show the legend
ax.legend()

# Display the plot
plt.show()
```

### 2. Do the spheres with equations
$$
x^2 + y^2 + z^2 = 1 \quad \text{and} \quad x^2 + y^2 + z^2 = 2
$$
have any common points?

### Solution:

To determine whether the spheres intersect, we need to analyze their equations. The first sphere is:

$$
x^2 + y^2 + z^2 = 1
$$

and the second sphere is:

$$
x^2 + y^2 + z^2 = 2
$$

Both spheres have the same center at \( (0, 0, 0) \), but their radii are different. The first sphere has a radius of:

$$
r_1 = \sqrt{1} = 1
$$

and the second sphere has a radius of:

$$
r_2 = \sqrt{2} \approx 1.414
$$

Since the radius of the second sphere is strictly larger than that of the first, the spheres cannot have any common points, as the first sphere is entirely inside the second one without intersecting it.

Thus, the spheres do not have any common points.

### Conclusion:

No, the spheres do not have any common points.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the spheres' parameters
center = np.array([0, 0, 0])  # Center at (0, 0, 0)

r1 = 1  # Radius of the first sphere (x^2 + y^2 + z^2 = 1)
r2 = np.sqrt(2)  # Radius of the second sphere (x^2 + y^2 + z^2 = 2)

# Create a meshgrid of spherical coordinates (theta and phi)
theta = np.linspace(0, np.pi, 100)  # Polar angle
phi = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle

# Create a grid of points in spherical coordinates
theta, phi = np.meshgrid(theta, phi)

# Parametrize the first sphere (radius r1)
x1 = center[0] + r1 * np.sin(theta) * np.cos(phi)
y1 = center[1] + r1 * np.sin(theta) * np.sin(phi)
z1 = center[2] + r1 * np.cos(theta)

# Parametrize the second sphere (radius r2)
x2 = center[0] + r2 * np.sin(theta) * np.cos(phi)
y2 = center[1] + r2 * np.sin(theta) * np.sin(phi)
z2 = center[2] + r2 * np.cos(theta)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the first sphere
ax.plot_surface(x1, y1, z1, color='cyan', alpha=0.6, label='Sphere 1: $x^2 + y^2 + z^2 = 1$')

# Plot the second sphere
ax.plot_surface(x2, y2, z2, color='orange', alpha=0.4, label='Sphere 2: $x^2 + y^2 + z^2 = 2$')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Spheres with Equations: $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$')

# Display the plot
plt.legend()
plt.show()
```

### 3. What curve in space is formed by the intersection of the spheres
$$
x^2 + y^2 + z^2 = 1 \quad \text{and} \quad (x-1)^2 + y^2 + z^2 = 1?
$$
Find the equation of this curve.

### Solution:

The first sphere has the equation:

$$
x^2 + y^2 + z^2 = 1
$$

The second sphere has the equation:

$$
(x-1)^2 + y^2 + z^2 = 1
$$

We now find the intersection of these two spheres. Expanding the equation of the second sphere:

$$
(x-1)^2 + y^2 + z^2 = 1
$$

Expanding the terms:

$$
(x^2 - 2x + 1) + y^2 + z^2 = 1
$$

Simplifying:

$$
x^2 + y^2 + z^2 - 2x + 1 = 1
$$

$$
x^2 + y^2 + z^2 - 2x = 0
$$

Now, subtract the equation of the first sphere, \( x^2 + y^2 + z^2 = 1 \), from this equation:

$$
(x^2 + y^2 + z^2 - 2x) - (x^2 + y^2 + z^2) = 0 - 1
$$

Simplifying:

$$
-2x = -1
$$

$$
x = \frac{1}{2}
$$

Substitute \( x = \frac{1}{2} \) into the equation of the first sphere:

$$
\left( \frac{1}{2} \right)^2 + y^2 + z^2 = 1
$$

Simplifying:

$$
\frac{1}{4} + y^2 + z^2 = 1
$$

$$
y^2 + z^2 = 1 - \frac{1}{4} = \frac{3}{4}
$$

Thus, the curve of intersection is a circle in the \( yz \)-plane with radius \( \frac{\sqrt{3}}{2} \), and the equation of the curve is:

$$
x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}
$$

### Conclusion:

The intersection curve is a circle in the \( yz \)-plane with radius \( \frac{\sqrt{3}}{2} \), and the equation of the curve is:

$$
x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the radius of the intersection circle
r_circle = np.sqrt(3)/2  # radius of the circle
x_value = 1/2  # x-coordinate of the intersection (x = 1/2)

# Create a meshgrid of angles for the circle
theta = np.linspace(0, 2 * np.pi, 100)  # full circle in the yz-plane

# Parametrize the circle in the yz-plane
y_circle = r_circle * np.cos(theta)
z_circle = r_circle * np.sin(theta)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the first sphere (x^2 + y^2 + z^2 = 1)
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x_sphere1 = np.outer(np.cos(u), np.sin(v))
y_sphere1 = np.outer(np.sin(u), np.sin(v))
z_sphere1 = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the second sphere ((x-1)^2 + y^2 + z^2 = 1)
x_sphere2 = np.outer(np.cos(u), np.sin(v)) + 1
y_sphere2 = np.outer(np.sin(u), np.sin(v))
z_sphere2 = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the first sphere
ax.plot_surface(x_sphere1, y_sphere1, z_sphere1, color='cyan', alpha=0.4, label=r'$x^2 + y^2 + z^2 = 1$')

# Plot the second sphere
ax.plot_surface(x_sphere2, y_sphere2, z_sphere2, color='orange', alpha=0.4, label=r'$(x-1)^2 + y^2 + z^2 = 1$')

# Plot the intersection circle at x = 1/2
ax.plot(y_circle, z_circle, np.full_like(y_circle, x_value), color='red', label=r'Intersection Curve: $x=\frac{1}{2}$')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Intersection of Two Spheres')

# Display the plot
plt.legend()
plt.show()
```

### 4. Write the equation of the tangent plane to the paraboloid
$$
z = (x - 1)^2 + y^2 + 1
$$
at the point \( P(1, 0, 1) \).

### Solution:

To find the equation of the tangent plane to a surface at a given point, we need the gradient of the surface. The equation of the paraboloid is:

$$
z = (x - 1)^2 + y^2 + 1
$$

We can express this as a function \( f(x, y, z) = z - (x - 1)^2 - y^2 - 1 \).

The gradient of \( f(x, y, z) \) is:

$$
\nabla f(x, y, z) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right)
$$

First, compute the partial derivatives:

$$
\frac{\partial f}{\partial x} = -2(x - 1), \quad \frac{\partial f}{\partial y} = -2y, \quad \frac{\partial f}{\partial z} = 1
$$

At the point \( P(1, 0, 1) \), the gradient is:

$$
\nabla f(1, 0, 1) = \left( -2(1 - 1), -2(0), 1 \right) = (0, 0, 1)
$$

Thus, the normal vector to the tangent plane is \( (0, 0, 1) \).

The equation of the tangent plane at \( P(1, 0, 1) \) is:

$$
0(x - 1) + 0(y - 0) + 1(z - 1) = 0
$$

Simplifying:

$$
z - 1 = 0
$$

Thus, the equation of the tangent plane is:

$$
z = 1
$$

### Conclusion:

The equation of the tangent plane to the paraboloid at the point \( P(1, 0, 1) \) is:

$$
z = 1
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the paraboloid equation: z = (x - 1)^2 + y^2 + 1
def paraboloid(x, y):
    return (x - 1)**2 + y**2 + 1

# Create a grid of x and y values
x = np.linspace(-2, 4, 100)
y = np.linspace(-2, 4, 100)
X, Y = np.meshgrid(x, y)
Z = paraboloid(X, Y)

# Create the tangent plane equation: z = 1
Z_tangent = np.ones_like(X)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the paraboloid surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7, rstride=100, cstride=100)

# Plot the tangent plane (z = 1)
ax.plot_surface(X, Y, Z_tangent, color='red', alpha=0.5, rstride=100, cstride=100)

# Mark the point P(1, 0, 1)
ax.scatter(1, 0, 1, color='black', s=100, label='Point P(1, 0, 1)', edgecolors='k')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Paraboloid and Tangent Plane at P(1, 0, 1)')

# Display the plot
plt.legend()
plt.show()
```
